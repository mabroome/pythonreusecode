```python
# The following function will normalize select columns in a dataset and return the dataset with the columns normalized
# dataset is the dataframe, sel_cols is a list of columns to be normalized ['abc', 'def' ...]
# Normalize will use standard (x - mean)/std deviation
# Other option is to change norm_type = "Minmax" and will normalize (x - min) / (max - min)

import pandas as pd
import numpy as np


def norm_vars(dataset, sel_cols, norm_type="Standard"):
    
    df1 = dataset.copy()
    df1.head()
    v = sel_cols

    df_norm = df1[v]
    df1.drop(v, axis=1, inplace=True)
    
    if norm_type == "Standard":

        from sklearn.preprocessing import StandardScaler

        # Establish object instance of MinMax Scalar

        sss = StandardScaler()

        # Normalize all columns : Output is an ndarry 

        var_norm = sss.fit_transform(df_norm)

        # Convert to dataframe

        df_norm = pd.DataFrame(data = var_norm, columns = df_norm.columns )
        
    elif norm_type == "MinMax":
        
        from sklearn.preprocessing import MinMaxScaler
        print("In MinMax")

        # Establish object instance of MinMax Scalar

        mms = MinMaxScaler()

        # Normalize all columns : Output is an ndarry 

        var_norm = mms.fit_transform(df_norm)
        
        # Convert to dataframe

        df_norm = pd.DataFrame(data = var_norm, columns = df_norm.columns )

    dataset = pd.concat([df1, df_norm], axis=1, sort=False)
    
    return dataset
```
