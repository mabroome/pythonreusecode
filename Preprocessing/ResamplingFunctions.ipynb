```python
def random_under_sample(df, y):
        
    from imblearn.under_sampling import RandomUnderSampler
    import pandas as pd
    import numpy as np
    
    rus = RandomUnderSampler(return_indices=False)
    df_sampled, y_sampled = rus.fit_sample(df, y)
    df_sampled = pd.DataFrame(df_sampled)
    
    return df_sampled, y_sampled

def random_over_sample(df, y):
        
    from imblearn.over_sampling import RandomOverSampler
    import pandas as pd
    import numpy as np
    
    ros = RandomOverSampler()
    df_sampled, y_sampled = ros.fit_sample(df, y)
    df_sampled = pd.DataFrame(df_sampled)
    
    return df_sampled, y_sampled

def SMOTE_sample(df, y):
    
    from imblearn.over_sampling import SMOTE
    import pandas as pd
    import numpy as np
    
    sm = SMOTE()
    x_sampled, y_sampled = sm.fit_sample(df, y)
    
    df_sampled = pd.DataFrame(x_sampled)
    
    return df_sampled, y_sampled

def ADASYN_sample(df, y):
 
    from imblearn.over_sampling import ADASYN 
    import pandas as pd
    import numpy as np
    
    ada = ADASYN()
    x_sampled, y_sampled = ada.fit_sample(df, y)
    
    df_sampled = pd.DataFrame(x_sampled)
    
    return df_sampled, y_sampled  
    
def SMOTEENN_sample(df, y):
 
    from imblearn.combine import SMOTEENN, SMOTETomek 
    import pandas as pd
    import numpy as np

    sm = SMOTEENN()
    x_sampled, y_sampled = sm.fit_sample(df, y)
    
    df_sampled = pd.DataFrame(x_sampled)
    
    return df_sampled, y_sampled


def SMOTETomek_sample(df, y):
 
    from imblearn.combine import SMOTEENN, SMOTETomek 
    import pandas as pd
    import numpy as np

    sm = SMOTETomek()
    x_sampled, y_sampled = sm.fit_sample(df, y)
    
    df_sampled = pd.DataFrame(x_sampled)
    
    return df_sampled, y_sampled
    
```
