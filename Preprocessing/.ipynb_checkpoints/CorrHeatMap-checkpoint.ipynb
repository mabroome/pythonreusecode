```python
# The funtion returns a correlation heat map
# include a pandas dataset and a list of columns 
# A correlation heat map is returned
```

```python
def corr_heat(dataset, columns, type='pearson'):
    # other correlation options include 'kengall' or 'spearman'
    # spearman is nonparametric
    
    import seaborn as sns
    
    df = dataset
    df_c = dataset[columns]
    
    var_corr = df_c.corr(type)
    
    sns.heatmap(var_corr, xticklabels=var_corr.columns, yticklabels=var_corr.columns, annot=True)
    
    
    
    return
```

```python
# Example of use is provided below
```

```python
import sklearn.datasets as ds
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

dataset = ds.load_boston()
dataset = pd.DataFrame(data = dataset['data'], columns = dataset['feature_names'])
# dataset.info()
dataset.head()

```

```python
# Select columns to be included in the correlation

columns = ["NOX", "INDUS", "DIS", "TAX", "LSTAT", "CRIM"]

# Call Function
corr_type = 'spearman' # 'pearson', 'spearman' or "kendall"

corr_heat(dataset, columns, corr_type)
```

```python
corr_type = 'pearson' # 'pearson', 'spearman' or "kendall"

corr_heat(dataset, columns, corr_type)
```

```python
corr_type = 'kendall' # 'pearson', 'spearman' or "kendall"

corr_heat(dataset, columns, corr_type)
```
